```python
import os
from os import listdir
from os.path import isfile, join

import codecs
import re

import urllib
```

```python
## get html file list
html_path = './corpus/html'
onlyfiles = [f for f in listdir(html_path) if isfile(join(html_path, f))]
print onlyfiles
```

```python
# solution 2
out_data  = 'ID\tAUTHOR\tALBUM\tPART\tMIDI PATH\tTEMPO\tDURATION\tDATE\tSIZE\tF0 PATH\tMP3 PATH\tOGG PATH\n'
out_idx = 1
for html_name in onlyfiles:
    full_path = html_path + '/' + html_name
    all_line  = ''

    f_h = codecs.open(full_path,'r','utf-8')
    for line in f_h:
        all_line += line.rstrip()
    f_h.close()
    
    author_name  = re.search('<h1 class.*>(.*)</h1>',all_line).group(1)
    print '\n' + author_name
    album_list = re.findall('<h2>.*?</table>',all_line)
    
    for one_album in album_list:
        song_set  = list()
        td_array  = list()
        # get album name
        album_name = re.search('<h2>(.*)</h2',one_album).group(1)
        print album_name
        # remove <th> row
        one_album = re.sub('<tr class=\"midi\">.*?</th></tr>','',one_album)
        # get song list 
        song_set = re.findall('<tr class=\"midi\">.*?</tr>',one_album)
        for each_song in song_set:
            # MIDI FILE, PART <td class="midi"><a href="midis/mendelssohn/mendel_op62_3.mid">No. 3 - Funeral march</a></td>
            # TEMPO           <td class="midi">Andante maestoso</td>
            # DURATION        <td class="midi" nowrap>3:06</td>
            # DATE            <td class="midi" nowrap>2014-04-03</td>
            # SIZE            <td class="midi" nowrap>9 KB</td>
            # F0              <td class="midi"><a class="navi" href="midis/mendelssohn/mendel_op62_3_format0.mid">?</a></td>
            # MP3             <td class="midi"><a href="mp3_link.php?id=886279303dc35f28650ca4f2e64c1b5d&amp;file=mendel_op62_3.mp3"><img alt="MP3" style="border-width:0" src="graphik/Loudspeaker.png"/></a></td>
            # OGG             <td class="midi"><a href="ogg/mendelssohn/mendel_op62_3.ogg"><img alt="OGG" style="border-width:0" src="graphik/Loudspeaker.png"></a></td>    
            
            #split raw data by <TD>           
            out_data += str(out_idx).zfill(5) + '\t' + author_name + '\t' + album_name
            out_idx += 1
            td_array = re.findall('<td class=\"midi\"\s*\w*>.*?</td>',each_song)
            #print '\n'.join(td_array)
            
            #----MIDI PATH, PART-----------------------------------------------
            match = re.search('href=\"(.*)\">(.*)</a>',td_array[0])
            rt = match.group(2)+'\t'+match.group(1) if match else 'None\tNone'
            out_data += '\t' + rt           
            #print 'midi_name, song_name ' + rt        
            #----TEMPO--------------------------------------------------------------
            match = re.search('<td.*>(.+)</td>',td_array[1])
            rt = match.group(1) if match else 'None'
            out_data += '\t' + rt
            #print 'tampo ' + rt            
            #----DURATION-----------------------------------------------------------
            match =  re.search('<td.*>(.+)</td>',td_array[2])
            rt = match.group(1) if match else 'None'
            out_data += '\t' + rt
            #print 'duration ' + rt             
            #----DATE --------------------------------------------------------------
            match =  re.search('<td.*>(.+)</td>',td_array[3])
            rt = match.group(1) if match else 'None'
            out_data += '\t' + rt
            #print 'date ' + rt             
            #----SIZE -----------------------------------------------------------
            match =  re.search('<td.*>(.+)</td>',td_array[4])
            rt = match.group(1) if match else 'None'
            out_data += '\t' + rt
            #print 'size ' + rt             
            #----FO-----------------------------------------------------------------
            match =  re.search('href=\"(.+)\"',td_array[5])
            rt = match.group(1) if match else 'None'
            out_data += '\t' + rt
            #print 'f0 ' + rt             
            #----MP3-----------------------------------------------------------------
            match =  re.search('href=\"(.*mp3)\"',td_array[6])
            rt = match.group(1) if match else 'None'
            out_data += '\t' + rt
            #print 'mp3 ' + rt             
            #----OGG-----------------------------------------------------------------
            match =  re.search('href=\"(.*ogg)\">',td_array[7])
            rt = match.group(1) if match else 'None'
            out_data += '\t' + rt
            #print 'ogg ' + rt
            #-----------------------------------------------------------------------
            out_data += '\n'
            #print out_data
      
    #break
#print out_data
f_h = codecs.open('midi_corpus.csv','w','utf-8')
f_h.write(out_data)
f_h.close()
```
